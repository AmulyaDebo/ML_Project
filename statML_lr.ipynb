```python
import dask.dataframe as dd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
import scipy.stats
from scipy.stats import chi2
%matplotlib inline
```

```python
import dask.array as da
df = pd.read_csv("/content/drive/MyDrive/new_train2.csv")
```

```python
df_test = pd.read_csv("/content/drive/MyDrive/new_test2.csv")
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df_test = df_test.loc[:, ~df_test.columns.str.contains('^Unnamed')]
```

```python
y=df["Label"]
df = df.drop("Label",axis=1)
X=df
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_t = std.fit_transform(X)
df_t = std.transform(df_test)
```

```python

import joblib
from dask.distributed import Client

client = Client(processes=False)
```

```python


from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs', max_iter=1000000, n_jobs=-1)
```

```python
with joblib.parallel_backend("dask"):
    lr.fit(X_t, y)
    
```

```python
y_pred_lr = lr.predict(df_t)
```

```python
predd_lr = pd.DataFrame(y_pred_lr, columns =["Label"])
predd_lr.to_csv('Prediction_lr1.csv', index=True)  
```

```python
lr2 = LogisticRegression(solver='lbfgs', max_iter=1000000, n_jobs=-1, class_weight='balanced')
```

```python

with joblib.parallel_backend("dask"):
    lr2.fit(X_t, y)
```

```python
y_pred_lr2 = lr2.predict(df_t)
```

```python
predd_lr2 = pd.DataFrame(y_pred_lr2, columns =["Label"])
predd_lr2.to_csv('Prediction_nlr2f3.csv', index=True)  
```
